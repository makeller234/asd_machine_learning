```python
# importing nessacary package

import pandas as pd
import warnings
warnings.simplefilter('ignore')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix as cm
import seaborn as sns
```

```python
# loading dataset
file = 'Resources/Autism.csv'
df = pd.read_csv(file)
df.head(5)
```

```python
# preprocessing the data 
```

```python
#reshape the data
X =df.drop(columns=["Class","User","Used_App_Before","Case No","Score","Residence","Screening Type","Language","Ethnicity"])
y =df[['Class']].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
df.head()
```

```python
# train, test, split
from sklearn.model_selection import train_test_split
```

```python
# use get dummies to turn catagorical data into 0 and 1 for machine learning
X = pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
pd.set_option('display.max_columns', None)
X_train.head()
```

```python
# check original df 
df.head()
```

```python
# y_train = pd.get_dummies(y_train)
```

# Logistic Regression

```python
# logistic regression

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
# using classifier for logitic regression

classifier.fit(X_train, y_train)
```

# We found a near perfect performance when running this model, which we could not explain at first. Looking deeper though we saw that our 10 of our inputs had a direct impact on the label. This led to a deeper understanding of the data that our project was based on.

```python
# checking training vs testing score

print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# checking how accuret predictions are

predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
# confusion matrix
#predictions, y_test
from sklearn.metrics import confusion_matrix as cm
predictions, y_test
cm = cm(y_test, predictions)
cm
```

# Confusion Matrix
this displayes the true positives, true negaitives and false positives and false negaitives.

```python
group_names = ["True Neg","False Pos","False Neg","True Pos"]

group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]

group_percentages = ["{0:.2%}".format(value) for value in cm.flatten()/np.sum(cm)]

labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
ax = sns.heatmap(cm, annot=labels, fmt="", )
ax.set_ylim(2,0)
```

# Feature Importance

```python
# find feature importance using classifier coeffiant
importance = classifier.coef_[0]

# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %s, Score: %.5f' % (X_train.columns[i],v))
    
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```

In this model we found that the "scores" were really a fill in for the label. You can see this as the 10 features with the biggest impact on the model were the ten questions on autism.
