Both Questions A5 and A6 have a 0.65 correlation to Score.  But their correlation to each other is only 0.44, meaning it's a weak correlation.  Which further means that we can't necessarily say that if someone answered question A5 a certain way, then they'll answer A6 in a certain way.  Ultimatley meaning, they're statistically likely to be independent variables, keep both inputs when fine-tuning feature selection.

Question A9 also has a 0.65 correlation to Score, and similar to above, it has weak correlation to both A5 (0.4) and A6 (0.45).  So again, we can statistically presume that A9 is an independent variable.

All of the AQ-10 question correlations to Score fall between 0.37 and 0.65.
